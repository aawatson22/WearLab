```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
```

```python
df = pd.DataFrame({
    2.75: [9,    7.5,  5.5,    3.75, 2.25],
    2.25: [8,    7,    5,      3.5,  2],
    1.75: [8,    7,    4.75,   3.5,  2],
    1.5:  [7.75, 6.75, 4.5,    3,    2],
    }, index=[8, 7, 6, 5, 4])
df
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
x = df.columns
y = df.index
x,y = np.meshgrid(x,y)
z = df
ax.scatter3D(x, y, z);

ax.set_xlabel("Width of Fabric")
ax.set_ylabel("Length of Fabric")
ax.set_zlabel("Change in Length")

ax.set_xticks(df.columns)
ax.tick_params(axis='y', rotation=-45, pad = -3 )

ax.tick_params(axis='x',  pad = 0 )

ax.figure.savefig('Fabric Max Stretch.eps', format='eps', dpi=1000)
```
