```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import seaborn as sns
%matplotlib inline
from sklearn.cross_validation import train_test_split
from sklearn import metrics 
from sklearn.linear_model import LinearRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score

from sklearn.linear_model import LinearRegression
import matplotlib.patches as mpatches
from sklearn.preprocessing import PolynomialFeatures
```

```python
Tall = pd.read_csv('Tall People.csv')
#Tall=Tall.dropna()
#Tall
```

```python
x=Tall['A2 Value']
y=Tall['angle']
x=x.reshape(-1,1)
y=y.reshape(-1,1)



polynomial_features= PolynomialFeatures(degree=2)
x_poly = polynomial_features.fit_transform(x)

model = LinearRegression()
tis=model.fit(x_poly, y)
predictions = model.predict(x_poly)


lm = LinearRegression()
lm.fit(x,y)
predictionslm = lm.predict(x)
plt.scatter(x,y, c='blue')
#x = x.reshape(-1,1)
plt.plot(x, model.predict(x_poly),color='k')
plt.xlabel("Distance Stretched (in)")
plt.ylabel("Resistance (kohms)")


print('Intercept:',model.intercept_)
print('Coefficient:',model.coef_)
#print('Mean Absolute Error:', metrics.mean_absolute_error(ytest, predictions))  
#print('Mean Squared Error:', metrics.mean_squared_error(ytest, predictions))  
#print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(ytest, predictions)))  
plt.savefig('Fabric.eps', format='eps', dpi=1000)
```
